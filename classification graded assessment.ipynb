```python
# importing required libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
```

### Loding the Dataset

```python
# reading the csv file and assigning it to a variable
electrical_grid = pd.read_csv("datasets/Data_for_UCI_named.csv")
```

```python
# loading the dataset
electrical_grid.head()
```

```python
# getting information about the dataset
electrical_grid.info()
```

```python
# getting a description of the dataset
electrical_grid.describe()
```

```python
# checking for empty fields in the dataset
electrical_grid.isnull().sum()
```

since there is no empty field in the dataset, we proceed further

```python
# dropping the "stab" column
new_electrical_grid_df = electrical_grid.drop(["stab"], axis = 1)
```

```python
# loading the "new_electrical_grid_df" dataframe
new_electrical_grid_df
```

```python
# encoding the categorival variable using Ordinal Encoder
ord_enc = OrdinalEncoder()
new_electrical_grid_df["stabf"] = ord_enc.fit_transform(new_electrical_grid_df[["stabf"]])
```

```python
# loading the encoded dataframe
new_electrical_grid_df
```

### Declaring the X and Y variables, Splitting and Scaling

```python
# declaring the X and y variables
X = new_electrical_grid_df.iloc[:, :-1].values
y = new_electrical_grid_df.iloc[:, -1].values

# splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

# carrying out scaling on the dataset before splitting
# applying the Standard Scaler technique to transform the 
# train set (x_train, y_train) and the test set (x_test)
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

# fitting Logistic Regression to the Training set
classifier = LogisticRegression(random_state = 1)
classifier.fit(X_train, y_train)

# predicting the Test set results
y_pred = classifier.predict(X_test)
```

### Evaluating the Logistic Regression Model

#### Confusuin Matrix

```python
# Making the Confusion Matrix    
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

we can see from the confusion matrix results, we have quite alot of correct predictions, in this case which is **1608** (494 + 1114) and **392** (174 + 218) incorrect predictions

### Using scikit learn to train a Random Forest and Extra Trees Classifier

#### Using Random Forest Classifier to train the model

```python
# fitting random forest classification to the Training set
rfc = RandomForestClassifier(random_state = 1)
rfc.fit(X_train, y_train)

# predicting the Test set results
rf_y_pred = classifier.predict(X_test)
```

##### Evaluating the Logistic Regression Model

```python
# Making the Confusion Matrix  
rfcm = confusion_matrix(y_test, rf_y_pred)
print(rfcm)
```

we can also see that from the confusion matrix results, we have quite alot of correct predictions, in this case which is **1608** (494 + 1114) and **392** (174 + 218) incorrect predictions

#### Using Extra Trees Classifier to train the model

```python
# fitting extra trees classification to the Training set
etc = ExtraTreesClassifier(n_estimators = 50, min_samples_split = 2, min_samples_leaf = 1, max_features = 'auto' )
etc.fit(X_train, y_train)

# predicting the Test set results
et_y_pred = etc.predict(X_test)
```

#### Evaluating the Logistic Regression Model

```python
# Making the Confusion Matrix  
etcm = confusion_matrix(y_test, et_y_pred)
print(etcm)
```

we can see from the confusion matrix of the Extra Tree Classifier, we have more correct predictions than the other two classifiers above. In this case we have **1857** (610 + 1247) correct predictions and **143** (41 + 102) incorrect predictions
